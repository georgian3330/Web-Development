## Markdown Syntax Overview

Here is a broad list of common Markdown syntaxes:

### 1. Headings
- `# Heading 1`
- `## Heading 2`
- `### Heading 3`
- `#### Heading 4`
- `##### Heading 5`
- `###### Heading 6`

### 2. Paragraphs and Line Breaks
- Plain text lines become paragraphs
- Line break: end a line with two spaces, or use a blank line between paragraphs

### 3. Emphasis
- Italic: `*italic*` or `_italic_`
- Bold: `**bold**` or `__bold__`
- Bold + italic: `***bold italic***`

### 4. Lists
- Unordered:
  - `- item`
  - `* item`
  - `+ item`
- Ordered:
  - `1. first`
  - `2. second`

### 5. Code
- Inline code: `` `code` ``
- Code block:
  - Fenced: 
    ````
    ```language
    code
    ```
    ````
  - Indented with 4 spaces or a tab

### 6. Blockquotes
- `> quote`
- Nested: `>> nested` or multiple `>`

### 7. Links
- Inline: `[text](url)`
- Reference: `[text][id]` + `[id]: url`
- Autolink: `<https://example.com>`
- Email: `<user@example.com>`

### 8. Images
- Inline: `![alt text](url)`
- Reference: `![alt][id]` + `[id]: url`

### 9. Horizontal Rules
- `---`
- `***`
- `___`

### 10. Tables (CommonMark / GFM)
- Basic:
  ```
  | Header | Header |
  | ------ | ------ |
  | Cell   | Cell   |
  ```

### 11. Task Lists (GitHub Flavored Markdown)
- `- [ ] incomplete`
- `- [x] complete`

### 12. Strikethrough (GFM)
- `~~strikethrough~~`

### 13. Footnotes (some Markdown variants)
- Inline: `Here is a note[^1]`
- Definition:
  ```
  [^1]: Footnote text
  ```

### 14. Definition Lists (some variants)
- Term:
  ```
  Term
  : Definition
  ```

### 15. HTML
- Raw HTML can be embedded directly in Markdown files
- Example: `<div>HTML block</div>`

### 16. Tables of Contents / Anchors (tool-specific)
- Some renderers support `[TOC]` or automatic anchor links

### 17. Emoji Shortcodes (GFM / specific renderers)
- `:smile:`

### 18. Math (Markdown extensions)
- Inline math: `$x^2$`
- Display math:
  ```
  $$ 
  E = mc^2
  $$
  ```

### 19. Definition of syntax extensions
- Certain platforms add custom syntax like admonitions, callouts, or embeds

> Note: Markdown has multiple dialects. The above list covers standard Markdown plus common GitHub-flavored extensions.
